```python
import numpy as np 
import pandas as pd
from sklearn import preprocessing
```

```python
data = np.genfromtxt("training_ccpp_x_y_train.csv",delimiter = ',')
df = pd.DataFrame(data)
df.describe()
```

```python
x = data[:,:4]
scaler = preprocessing.StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
df2 = pd.DataFrame(x)
df2.head()
df2['C'] = 1
df2.head()
points = df2.values
points
```

```python
def cost(points,m):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i,:5]
        y = points[i,-1]
        for j in range(5):
            total_cost += (-2/M)*(y - (m*x).sum())**2
        return total_cost

def stepgradient(points,learning_rate,m):
    m_slope = np.zeros(5)
    M = len(points)
    for i in range(M):
        x = points[i,:5]
        y = points[i,-1]
        for j in range(5):
            m_slope[j] += (-2/M)*(y - (m*x).sum())*x[j]
        new_m = m - learning_rate*m_slope
    return new_m

def gd(points,learning_rate,number_of_iteration):
    m = np.zeros(5)
    for i in range(number_of_iteration):
        m = stepgradient(points,learning_rate,m)
        print(i , "Cost :", cost(points,m))
    return m

def run():
    data = np.genfromtxt("training_ccpp_x_y_train.csv",delimiter = ',')
    x = data[:,:4]
    scaler = preprocessing.StandardScaler()
    scaler.fit(x)
    x = scaler.transform(x)
    df2 = pd.DataFrame(x)
    df2['C'] = 1
    points = df2.values
    learning_rate = 0.001
    number_of_iteration = 100
    m = gd(points,learning_rate,number_of_iteration)
    return m,scaler

def pred():
    m,v = run()
    data2 = np.genfromtxt("test_ccpp_x_test.csv",delimiter = ',')
    l = v.transform(data2)
    df2 = pd.DataFrame(l)
    df2['C'] = 1
    points = df2.values
    M = len(points)
    y = np.zeros(M)
    for i in range(M):
        x = points[i,:5]
        y[i] += (m*x).sum()
    np.savetxt("predict.csv",y,fmt='%0.5f')
```
