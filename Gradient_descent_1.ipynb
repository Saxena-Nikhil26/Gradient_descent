```python
import numpy as np 
import pandas as pd
from sklearn import preprocessing
```

```python
data = np.genfromtxt("training_ccpp_x_y_train.csv",delimiter = ',')
x = data[:,:4]
y = data[:,4]
x
```

```python
def cost(points,Y,m):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i,:]
        y = Y[i]
        for j in range(5):
            total_cost += (-2/M)*(y - (m*x).sum())**2
        return total_cost

def stepgradient(points, y,learning_rate,m):
    m_slope = np.zeros(len(m))
    M, N = points.shape
    for i in range(M):
        x = points[i,:]
        Y = y[i]
        for j in range(5):
            m_slope[j] += (-2/M)*(Y - (m*x).sum())*x[j]
        new_m = m - learning_rate*m_slope
    return new_m

def gd(points,Y,learning_rate,number_of_iteration):
    m = np.zeros(5)
    for i in range(number_of_iteration):
        m = stepgradient(points,Y,learning_rate,m)
        print(i , "Cost :", cost(points,Y,m))
    return m

scaler = preprocessing.StandardScaler()
    
def run():
    data = np.genfromtxt("training_ccpp_x_y_train.csv",delimiter = ',')
    x = data[:,:-1]
    y = data[:,-1]
    scaler.fit(x)
    x = scaler.transform(x)
    M,N = data.shape
    x = np.append(x, np.ones([M,1]), axis = 1)
    learning_rate = 0.2
    number_of_iteration = 100
    m = gd(x, y,learning_rate,number_of_iteration)
    return m

def pred():
    m = run()
    points2 = np.genfromtxt("test_ccpp_x_test.csv",delimiter = ',')
    points2 = scaler.transform(points2)
    points2 = np.append(points2, np.ones([len(points2),1]), axis=1)
    M = len(points2)
    y = np.zeros(M)
    for i in range(M):
        x = points2[i,:]
        y[i] += (m*x).sum()
    np.savetxt("predict.csv",y,delimiter = ',')
```
